How to concatenate dataframes

```python
first10 = surveys.head(10)
first10
```

```python
last10 = surveys.tail(11)
last10 = last10.reset_index(drop=True)
last10
```

```python
vertical = pd.concat([first10, last10], axis=0)
vertical
```

```python
vertical.to_csv('out.csv', index=False)
```

```python
pd.read_csv('out.csv')
```

```python
surveys2001 = surveys[surveys['year'] == 2001]
surveys2002 = surveys[surveys['year'] == 2002]
```

```python
all_data = pd.concat([surveys2001, surveys2002], axis=0)
```

```python
weight_year = all_data.groupby(['year', 'sex'])['weight'].mean().unstack()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
weight_year.plot(kind='bar', stacked=True)
```

```python
weight_year.to_csv('w.csv')
```

Inner join

```python
species = pd.read_csv('data/speciesSubset.csv')
species
```

```python
surveys_sub = surveys.head(10)
surveys_sub
```

```python
merged_inner = pd.merge(left=surveys_sub, right=species,
                       left_on='species_id', right_on='species_id')
merged_inner.shape
```

```python
merged_inner
```

```python
merged_left = pd.merge(left=surveys_sub, right=species,
                       how='left', on='species_id')
merged_left.shape
```

```python
merged_left
```
