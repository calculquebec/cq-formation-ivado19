```python
grouped_plot_data = small_dataset.groupby('sex')
colors = ['r', 'g']
color_index = 0
for group in grouped_plot_data: 
    color = colors[color_index]
    group_label = group[0]
    group_data = group[1]
    plt.plot(group_data['plot_id'], color = color, label = group_label)
    
    color_index  = color_index + 1

plt.legend()
```
