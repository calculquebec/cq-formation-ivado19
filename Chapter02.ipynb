### Summary Example

```python
by_site_sex = surveys_df.groupby(['plot_id', 'sex'])
sum_weights = by_site_sex['weight'].sum()
swu = sum_weights.unstack()
```

```python
swu.plot?
```

```python
s_plot = swu.plot(kind='bar', stacked=True,
                  title="Sum weights for each site")
s_plot.set_ylabel("Weight")
s_plot.set_xlabel("Site")
```
